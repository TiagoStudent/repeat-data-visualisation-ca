```python
# ── Library Imports ──────────────────────────────────────────────────────────
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
import plotly
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import ipywidgets as widgets
from IPython.display import display, HTML
import warnings
warnings.filterwarnings('ignore')

# Visual style
sns.set_theme(style='whitegrid', palette='muted')
plt.rcParams.update({'figure.dpi': 120, 'font.size': 11})

print("Libraries loaded successfully.")
print(f"pandas {pd.__version__} | numpy {np.__version__} | plotly {plotly.__version__}")
```

```python
# ── 1.1 Load the dataset ─────────────────────────────────────────────────────
import os, pathlib, glob
# Search for the dataset in the current working directory and common locations
# Search for the dataset — supports both 'OnlineRetail.xlsx' and 'Online Retail.xlsx'
_found = glob.glob(str(pathlib.Path(os.getcwd()) / '**' / '*etail*.xlsx'), recursive=True)
if not _found:
    raise FileNotFoundError(
        f'Dataset not found. Current directory: {os.getcwd()}\n'
        'Please place OnlineRetail.xlsx (or Online Retail.xlsx) in the same folder as this notebook.')
_data_path = pathlib.Path(_found[0])
print(f'Found dataset at: {_data_path.resolve()}')
df_raw = pd.read_excel(_data_path)

print(f"Dataset shape: {df_raw.shape[0]:,} rows × {df_raw.shape[1]} columns")
print("\nColumn names and data types:")
print(df_raw.dtypes)
print("\nFirst 5 rows:")
df_raw.head()
```

```python
# ── 1.2 Missing Values ───────────────────────────────────────────────────────
missing = df_raw.isnull().sum()
missing_pct = (missing / len(df_raw) * 100).round(2)
missing_df = pd.DataFrame({'Missing Count': missing, 'Missing %': missing_pct})
missing_df = missing_df[missing_df['Missing Count'] > 0]
print("Columns with missing values:")
print(missing_df)
```

```python
# ── 1.3 Identify data quality issues ─────────────────────────────────────────
cancelled = df_raw[df_raw['InvoiceNo'].astype(str).str.startswith('C')]
neg_qty   = df_raw[df_raw['Quantity'] <= 0]
bad_price = df_raw[df_raw['UnitPrice'] <= 0]
dupes     = df_raw[df_raw.duplicated()]

print(f"Cancelled transactions (InvoiceNo starts with 'C') : {len(cancelled):>7,}")
print(f"Rows with Quantity <= 0                            : {len(neg_qty):>7,}")
print(f"Rows with UnitPrice <= 0                           : {len(bad_price):>7,}")
print(f"Duplicate rows                                     : {len(dupes):>7,}")
```

```python
# ── 1.4 Apply cleaning steps ─────────────────────────────────────────────────
df = df_raw.copy()

# Step 1: Remove exact duplicates
before = len(df)
df = df.drop_duplicates()
print(f"Step 1 — Remove duplicates        : {before - len(df):>6,} rows removed | Remaining: {len(df):,}")

# Step 2: Remove cancelled transactions
before = len(df)
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
print(f"Step 2 — Remove cancellations      : {before - len(df):>6,} rows removed | Remaining: {len(df):,}")

# Step 3: Remove non-positive quantities
before = len(df)
df = df[df['Quantity'] > 0]
print(f"Step 3 — Remove Quantity <= 0      : {before - len(df):>6,} rows removed | Remaining: {len(df):,}")

# Step 4: Remove non-positive unit prices
before = len(df)
df = df[df['UnitPrice'] > 0]
print(f"Step 4 — Remove UnitPrice <= 0     : {before - len(df):>6,} rows removed | Remaining: {len(df):,}")

# Step 5: Drop rows with missing Description
before = len(df)
df = df.dropna(subset=['Description'])
print(f"Step 5 — Drop missing Description  : {before - len(df):>6,} rows removed | Remaining: {len(df):,}")

print(f"\nFinal cleaned dataset: {len(df):,} rows ({len(df)/len(df_raw)*100:.1f}% of original)")
```

```python
# ── 1.5 Feature engineering ──────────────────────────────────────────────────

# Flag for customer-level analysis
df['HasCustomerID'] = df['CustomerID'].notna()

# Derived variable: total line value
df['TotalValue'] = df['Quantity'] * df['UnitPrice']

# Date components
df['Year']       = df['InvoiceDate'].dt.year
df['Month']      = df['InvoiceDate'].dt.month
df['DayOfWeek']  = df['InvoiceDate'].dt.day_name()
df['Hour']       = df['InvoiceDate'].dt.hour
df['YearMonth']  = df['InvoiceDate'].dt.to_period('M')

print("Derived variables created:")
print(df[['InvoiceDate','TotalValue','Year','Month','DayOfWeek','Hour','YearMonth']].head(3))
print(f"\nTotalValue range: £{df['TotalValue'].min():.2f} – £{df['TotalValue'].max():,.2f}")
print(f"Mean TotalValue per line: £{df['TotalValue'].mean():.2f}")
```
