```python
# ── Library Imports ──────────────────────────────────────────────────────────
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
import plotly
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import ipywidgets as widgets
from IPython.display import display, HTML
import warnings
warnings.filterwarnings('ignore')

# Visual style
sns.set_theme(style='whitegrid', palette='muted')
plt.rcParams.update({'figure.dpi': 120, 'font.size': 11})

print("Libraries loaded successfully.")
print(f"pandas {pd.__version__} | numpy {np.__version__} | plotly {plotly.__version__}")
```

```python
# ── 1.1 Load the dataset ─────────────────────────────────────────────────────
import os, pathlib, glob
# Search for the dataset in the current working directory and common locations
# Search for the dataset — supports both 'OnlineRetail.xlsx' and 'Online Retail.xlsx'
_found = glob.glob(str(pathlib.Path(os.getcwd()) / '**' / '*etail*.xlsx'), recursive=True)
if not _found:
    raise FileNotFoundError(
        f'Dataset not found. Current directory: {os.getcwd()}\n'
        'Please place OnlineRetail.xlsx (or Online Retail.xlsx) in the same folder as this notebook.')
_data_path = pathlib.Path(_found[0])
print(f'Found dataset at: {_data_path.resolve()}')
df_raw = pd.read_excel(_data_path)

print(f"Dataset shape: {df_raw.shape[0]:,} rows × {df_raw.shape[1]} columns")
print("\nColumn names and data types:")
print(df_raw.dtypes)
print("\nFirst 5 rows:")
df_raw.head()
```

```python
# ── 1.2 Missing Values ───────────────────────────────────────────────────────
missing = df_raw.isnull().sum()
missing_pct = (missing / len(df_raw) * 100).round(2)
missing_df = pd.DataFrame({'Missing Count': missing, 'Missing %': missing_pct})
missing_df = missing_df[missing_df['Missing Count'] > 0]
print("Columns with missing values:")
print(missing_df)
```

```python
# ── 1.3 Identify data quality issues ─────────────────────────────────────────
cancelled = df_raw[df_raw['InvoiceNo'].astype(str).str.startswith('C')]
neg_qty   = df_raw[df_raw['Quantity'] <= 0]
bad_price = df_raw[df_raw['UnitPrice'] <= 0]
dupes     = df_raw[df_raw.duplicated()]

print(f"Cancelled transactions (InvoiceNo starts with 'C') : {len(cancelled):>7,}")
print(f"Rows with Quantity <= 0                            : {len(neg_qty):>7,}")
print(f"Rows with UnitPrice <= 0                           : {len(bad_price):>7,}")
print(f"Duplicate rows                                     : {len(dupes):>7,}")
```

```python
# ── 1.4 Apply cleaning steps ─────────────────────────────────────────────────
df = df_raw.copy()

# Step 1: Remove exact duplicates
before = len(df)
df = df.drop_duplicates()
print(f"Step 1 — Remove duplicates        : {before - len(df):>6,} rows removed | Remaining: {len(df):,}")

# Step 2: Remove cancelled transactions
before = len(df)
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
print(f"Step 2 — Remove cancellations      : {before - len(df):>6,} rows removed | Remaining: {len(df):,}")

# Step 3: Remove non-positive quantities
before = len(df)
df = df[df['Quantity'] > 0]
print(f"Step 3 — Remove Quantity <= 0      : {before - len(df):>6,} rows removed | Remaining: {len(df):,}")

# Step 4: Remove non-positive unit prices
before = len(df)
df = df[df['UnitPrice'] > 0]
print(f"Step 4 — Remove UnitPrice <= 0     : {before - len(df):>6,} rows removed | Remaining: {len(df):,}")

# Step 5: Drop rows with missing Description
before = len(df)
df = df.dropna(subset=['Description'])
print(f"Step 5 — Drop missing Description  : {before - len(df):>6,} rows removed | Remaining: {len(df):,}")

print(f"\nFinal cleaned dataset: {len(df):,} rows ({len(df)/len(df_raw)*100:.1f}% of original)")
```

```python
# ── 1.5 Feature engineering ──────────────────────────────────────────────────

# Flag for customer-level analysis
df['HasCustomerID'] = df['CustomerID'].notna()

# Derived variable: total line value
df['TotalValue'] = df['Quantity'] * df['UnitPrice']

# Date components
df['Year']       = df['InvoiceDate'].dt.year
df['Month']      = df['InvoiceDate'].dt.month
df['DayOfWeek']  = df['InvoiceDate'].dt.day_name()
df['Hour']       = df['InvoiceDate'].dt.hour
df['YearMonth']  = df['InvoiceDate'].dt.to_period('M')

print("Derived variables created:")
print(df[['InvoiceDate','TotalValue','Year','Month','DayOfWeek','Hour','YearMonth']].head(3))
print(f"\nTotalValue range: £{df['TotalValue'].min():.2f} – £{df['TotalValue'].max():,.2f}")
print(f"Mean TotalValue per line: £{df['TotalValue'].mean():.2f}")
```

```python
# ── 2.1 Descriptive Statistics ───────────────────────────────────────────────
print("=== Descriptive Statistics ===\n")
print(f"Date range     : {df['InvoiceDate'].min().date()} to {df['InvoiceDate'].max().date()}")
print(f"Total revenue  : £{df['TotalValue'].sum():>12,.2f}")
print(f"Unique invoices: {df['InvoiceNo'].nunique():>12,}")
print(f"Unique products: {df['StockCode'].nunique():>12,}")
print(f"Unique customers (with ID): {int(df['CustomerID'].nunique()):>6,}")
print(f"Countries      : {df['Country'].nunique():>12,}")
print()
print(df[['Quantity','UnitPrice','TotalValue']].describe().round(2))
```

```python
# ── 2.2 Monthly Revenue (Line Chart) ─────────────────────────────────────────
monthly = df.groupby('YearMonth')['TotalValue'].sum().reset_index()
monthly['YearMonth_str'] = monthly['YearMonth'].astype(str)

fig, ax = plt.subplots(figsize=(13, 4.5))
ax.plot(monthly['YearMonth_str'], monthly['TotalValue'] / 1e6,
        marker='o', linewidth=2.5, color='steelblue', markersize=7)
ax.fill_between(monthly['YearMonth_str'], monthly['TotalValue'] / 1e6,
                alpha=0.12, color='steelblue')
ax.set_title('Monthly Revenue — Dec 2010 to Dec 2011', fontsize=14, fontweight='bold', pad=12)
ax.set_xlabel('Month', fontsize=11)
ax.set_ylabel('Revenue (£ millions)', fontsize=11)
ax.yaxis.set_major_formatter(mticker.FormatStrFormatter('£%.1fM'))
plt.xticks(rotation=40, ha='right', fontsize=9)
plt.tight_layout()
plt.show()
```

```python
# ── 2.3 Top 10 Products by Revenue (Horizontal Bar) ─────────────────────────
top_products = (df.groupby('Description')['TotalValue']
                  .sum()
                  .sort_values(ascending=False)
                  .head(10)
                  .reset_index())

fig, ax = plt.subplots(figsize=(11, 5.5))
bars = ax.barh(top_products['Description'][::-1],
               top_products['TotalValue'][::-1] / 1e3,
               color='teal', edgecolor='white', linewidth=0.5)
ax.set_title('Top 10 Products by Revenue', fontsize=14, fontweight='bold', pad=12)
ax.set_xlabel('Revenue (£ thousands)', fontsize=11)
ax.xaxis.set_major_formatter(mticker.FormatStrFormatter('£%.0fK'))
for bar in bars:
    ax.text(bar.get_width() + 1, bar.get_y() + bar.get_height()/2,
            f'£{bar.get_width():.0f}K', va='center', fontsize=9)
sns.despine(left=True, bottom=False)
plt.tight_layout()
plt.show()
```

```python
# ── 2.4 Top 10 Countries by Revenue (Bar Chart) ─────────────────────────────
top_countries = (df.groupby('Country')['TotalValue']
                   .sum()
                   .sort_values(ascending=False)
                   .head(10)
                   .reset_index())

colors = ['#1565C0' if c == 'United Kingdom' else '#64B5F6'
          for c in top_countries['Country']]

fig, ax = plt.subplots(figsize=(11, 5))
ax.bar(top_countries['Country'], top_countries['TotalValue'] / 1e6,
       color=colors, edgecolor='white', linewidth=0.5)
ax.set_title('Top 10 Countries by Revenue', fontsize=14, fontweight='bold', pad=12)
ax.set_xlabel('Country', fontsize=11)
ax.set_ylabel('Revenue (£ millions)', fontsize=11)
ax.yaxis.set_major_formatter(mticker.FormatStrFormatter('£%.1fM'))
plt.xticks(rotation=30, ha='right', fontsize=10)
plt.tight_layout()
plt.show()
```

```python
# ── 2.5 Distribution of Order Values (Histogram) ────────────────────────────
order_vals = df.groupby('InvoiceNo')['TotalValue'].sum()
order_clipped = order_vals[order_vals <= 1000]

fig, axes = plt.subplots(1, 2, figsize=(13, 4.5))

# Left: full distribution clipped at £1,000
axes[0].hist(order_clipped, bins=60, color='coral', edgecolor='white', linewidth=0.4)
axes[0].set_title('Order Value Distribution (≤ £1,000)', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Order Value (£)')
axes[0].set_ylabel('Number of Orders')
axes[0].xaxis.set_major_formatter(mticker.FormatStrFormatter('£%.0f'))

# Right: log-scale to show full range
axes[1].hist(order_vals, bins=80, color='coral', edgecolor='white', linewidth=0.4)
axes[1].set_yscale('log')
axes[1].set_title('Order Value Distribution (log scale, all orders)', fontsize=12, fontweight='bold')
axes[1].set_xlabel('Order Value (£)')
axes[1].set_ylabel('Number of Orders (log scale)')
axes[1].xaxis.set_major_formatter(mticker.FormatStrFormatter('£%.0f'))

plt.tight_layout()
plt.show()
```

```python
# ── 2.6 Revenue by Day of Week (Bar Chart) ───────────────────────────────────
dow_order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
dow = (df.groupby('DayOfWeek')['TotalValue']
         .sum()
         .reindex(dow_order)
         .reset_index())

colors_dow = ['#42A5F5' if d not in ['Saturday','Sunday'] else '#EF9A9A'
              for d in dow['DayOfWeek']]

fig, ax = plt.subplots(figsize=(9, 4.5))
ax.bar(dow['DayOfWeek'], dow['TotalValue'] / 1e6,
       color=colors_dow, edgecolor='white', linewidth=0.5)
ax.set_title('Total Revenue by Day of Week', fontsize=14, fontweight='bold', pad=12)
ax.set_xlabel('Day of Week', fontsize=11)
ax.set_ylabel('Revenue (£ millions)', fontsize=11)
ax.yaxis.set_major_formatter(mticker.FormatStrFormatter('£%.1fM'))
plt.tight_layout()
plt.show()
```

```python
# ── 2.7 Quantity vs Unit Price Scatter ───────────────────────────────────────
sample = df[(df['UnitPrice'] <= 20) & (df['Quantity'] <= 150)].sample(4000, random_state=42)

fig, ax = plt.subplots(figsize=(9, 5))
scatter = ax.scatter(sample['UnitPrice'], sample['Quantity'],
                     c=sample['TotalValue'], cmap='YlOrRd',
                     alpha=0.45, s=18, edgecolors='none')
cbar = plt.colorbar(scatter, ax=ax)
cbar.set_label('Line Value (£)', fontsize=10)
ax.set_title('Quantity vs Unit Price (sample of 4,000 transactions)', fontsize=13, fontweight='bold')
ax.set_xlabel('Unit Price (£)', fontsize=11)
ax.set_ylabel('Quantity Ordered', fontsize=11)
plt.tight_layout()
plt.show()
```
